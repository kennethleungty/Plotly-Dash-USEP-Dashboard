```python
import numpy as np 
import pandas as pd
import os
from sklearn import metrics
from matplotlib import pyplot as plt
```

```python
print(os.getcwd())
os.chdir('/Users/nikhilthomas/Desktop/EMc/Full Data Sets/')
print(os.getcwd())
```

```python
# read files 

# Novemeber Forecast 
        #Test_start = '2019-11-08'
        #Test_end = '2019-11-15'
DF_nov = pd.read_csv('results_nov.csv')

# Novemeber Forecast 
        #Test_start = '2019-12-06'
        #Test_end = '2019-12-13'
DF_dec = pd.read_csv('results_dec.csv')
```

```python
DF_nov = DF_nov[['TRADING_DATE', 'PERIOD', 'USEP','ensemble_1']]
DF_nov.columns = ['TRADING_DATE', 'PERIOD', 'Actuals','Forecast_Baseline']
DF_nov['User_mod']=1
DF_nov['Forecast_mod']=DF_nov['User_mod'] * DF_nov['Forecast_Baseline']

DF_dec = DF_dec[['TRADING_DATE', 'PERIOD', 'USEP','ensemble_1']]
DF_dec.columns = ['TRADING_DATE', 'PERIOD', 'Actuals','Forecast_Baseline']
DF_dec['User_mod']=1
DF_dec['Forecast_mod']=DF_dec['User_mod'] * DF_dec['Forecast_Baseline']
```

```python
DF_FINAL = DF_nov.append(DF_dec)
```

# Plotting

#### Filter data based on input

```python
Date_filter = '2019-11-13'
```

```python
DF_filter = DF_FINAL[DF_FINAL.TRADING_DATE == Date_filter]
```

### Baseline Plot

```python
plt.plot(DF_filter.PERIOD,DF_filter.Forecast_Baseline, label = 'Baseline')

plt.xlabel('Period')
plt.ylabel('USEP')
plt.title('Baseline Forecast of USEP '+str(Date_filter)+' ')
plt.legend()
plt.show()
```

### Modified Forecast Plot

#### Obtain user input on modification 

```python
DF_filter.User_mod = 0.9
```

#### Calculate mod forecast

```python
DF_filter['Forecast_mod']=DF_filter['User_mod'] * DF_filter['Forecast_Baseline']
```

```python
plt.plot(DF_filter.PERIOD,DF_filter.Forecast_Baseline, label = 'Baseline')
plt.plot(DF_filter.PERIOD,DF_filter.Forecast_mod, label = 'Modified')
plt.xlabel('Period')
plt.ylabel('USEP')
plt.title('Baseline Forecast of USEP '+str(Date_filter)+' ')
plt.legend()
plt.show()
```

### All forecasts + Actuals plot


```python
plt.plot(DF_filter.PERIOD,DF_filter.Forecast_Baseline, label = 'Baseline')
plt.plot(DF_filter.PERIOD,DF_filter.Forecast_mod, label = 'Modified')
plt.plot(DF_filter.PERIOD,DF_filter.Actuals, label = 'Actuals')


plt.xlabel('Period')
plt.ylabel('USEP')
plt.title('Baseline Forecast of USEP '+str(Date_filter)+' ')
plt.legend()
plt.show()
```

```python

print('Baseline Accuracy Measure') 
print('MAE is: ',metrics.mean_absolute_error(DF_filter.Actuals, DF_filter.Forecast_Baseline))
print('RMSE is: ',np.sqrt(metrics.mean_squared_error(DF_filter.Actuals, DF_filter.Forecast_Baseline)))

print('Modified Forecast Accuracy Measure')
print('MAE is: ',metrics.mean_absolute_error(DF_filter.Actuals, DF_filter.Forecast_mod))
print('RMSE is: ',np.sqrt(metrics.mean_squared_error(DF_filter.Actuals, DF_filter.Forecast_mod)))


```
